```python
#Linear Regrassion
lrg = LinearRegression()

lrg.fit(x_train,y_train)

lrg.score(x_train,y_train)*100
```

```python
x_test_pred_dt = dt.predict(x_test)
test_pred_dt = accuracy_score(x_test_pred_dt, y_test)
print("Decision Tree test : ", test_pred_dt *100)
```

```python
y_pred_dt = dt.predict(x_test)
print(classification_report(y_test, y_pred_dt))
```
